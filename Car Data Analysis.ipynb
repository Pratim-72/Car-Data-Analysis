```python
# 1. Linear Regression

data['PowerTrain'].replace(to_replace = ['RWD' , 'FWD' , 'AWD'] , value = [0 , 1 , 2] , inplace = True)

X = data[['AccelSec' , 'TopSpeed_KmH' , 'Range_Km' , 'Efficiency_WhKm' , 'RapidCharge' , 'PowerTrain']]
Y = data['INR']
```

```python
# Finding linear regression using ols method

import statsmodels.api as sm

X = sm.add_constant(X)
results = sm.OLS(Y , X)
```

```python
models = results.fit()
models.summary()
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train , Y_train)
pred = lr.predict(X_test)
```

```python
# R2 Score

from sklearn.metrics import accuracy_score , r2_score

r2_scoree = r2_score(Y_test , pred)
r2_scoree *100
```

```python
# Around 53% of the dependant variable has been explained by the independant variables
```

```python
# Applying PCA for dimension reduction

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

x = ['AccelSec' , 'TopSpeed_KmH' , 'Range_Km' , 'Efficiency_WhKm' , 'FastCharge_KmH' , 'RapidCharge' , 'PowerTrain' , 'Seats' , 'INR']

x = data.loc[: , x].values
x = StandardScaler().fit_transform(x)

```

```python
pca = PCA(n_components = 9)

s = pca.fit_transform(x)

data1 = pd.DataFrame(s, columns=['PC1', 'PC2','PC3','PC4','Pc5','PC6', 'PC7', 'PC8','PC9'])
data1
```

```python
data1.head()
```

```python
# Finding the correlation coefficient between the original variable and the components

loadings = pca.components_
features = pca.n_features_
pc = ['PC' + str(i) for i in list(range(1 , features + 1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc , loadings)))
loadings_df['Variable'] = data1.columns.values
loadings_df = loadings_df.set_index('Variable')
loadings_df
```

```python
# Correlation matrix through heatmap
plt.rcParams['figure.figsize'] = (20 , 15)
sn.heatmap(loadings_df , annot = True , cmap = 'Spectral')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage , dendrogram

linkage = linkage(data1 , 'complete')
plt.figure(figsize = (13 , 9))
dendrogram(linkage , orientation = 'top')
plt.show()
```

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values , pca.explained_variance_ratio_ , 'o-' , linewidth = 2 , color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 4 , random_state = 1234)

kmeans.fit(s)

labels = kmeans.labels_

centroid = kmeans.cluster_centers_
```

```python
print('labels')
print(labels)
print()
print('centers of cluster')
print(centroid)
```
